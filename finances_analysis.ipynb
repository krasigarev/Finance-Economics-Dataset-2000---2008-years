## Visualizations

```python
numeric_df = df.select_dtypes(include=[np.number])

if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(14,9))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.show()
else:
    print('Not enough numeric columns for a correlation heatmap.')
```

```python
df.columns
```

####  Select only numeric columns for correlation heatmap

```python
selected_columns = ['Open Price', 'Close Price', 'Daily High', 'Daily Low']

if all(col in df.columns for col in selected_columns):
    sns.pairplot(df[selected_columns].dropna())
    plt.suptitle('Pair Plot of Selected Financial Indicators', y=1.02)
    plt.show()
else:
    print('Some selected columns for pair plot are missing.')
```

#### Distribution of Trading Volume

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Trading Volume'].dropna(), bins=30, kde=True)
plt.title('Distribution of Trading Volume')
plt.xlabel('Trading Volume')
plt.ylabel('Frequency')
plt.show()
```

#### Bar Plot using a count plot for Stock Index frequency

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Stock Index', order=df['Stock Index'].value_counts().index, palette='bright')
plt.title('Frequency of Stock Indexes')
plt.xlabel('Stock Index')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Distribution of Real Estate Index

```python
plt.figure(figsize=(14, 9))
sns.histplot(df['Real Estate Index'].dropna(), bins=30, color='blue', kde=True, line_kws={'color': 'crimson', 'lw': 6, 'ls': '--'})
plt.title('Distribution of Real Estate Index')
plt.xlabel('Real Estate Index')
plt.ylabel('Count')
plt.show()
```

### Distribution of Consumer Confidence Index

```python
sns.displot(data=df, x='Date', y='Consumer Confidence Index', cbar=True)
plt.xticks(rotation=45)
plt.title('Distribution of Consumer Confidence Index')
plt.show()
```

### Relationship between Inflation and Unemployment. Percent at (%).

```python
df.groupby('Stock Index')[['Inflation Rate (%)', 'Unemployment Rate (%)']].aggregate({'Inflation Rate (%)':['mean', 'std'], 'Unemployment Rate (%)':['mean', 'std']})
```

```python
df.groupby('Stock Index')[['Inflation Rate (%)', 'Unemployment Rate (%)']].aggregate({'Inflation Rate (%)':['mean', 'std'], 'Unemployment Rate (%)':['mean', 'std']}).plot(kind='barh')
plt.yticks(rotation=45)
plt.title('Relationship between Inflation and Unemployment');
```

### Relationship between Inflation Rate and Consumer Confidence Index. Percent at (%)

```python
df.groupby('Stock Index')[['Inflation Rate (%)', 'Consumer Confidence Index']].agg({'Inflation Rate (%)': ['min', 'max', 'mean', 'std'], 'Consumer Confidence Index':['min', 'max', 'mean', 'std']})
```

```python
df.groupby('Stock Index')[['Inflation Rate (%)', 'Consumer Confidence Index']].agg({'Inflation Rate (%)': ['min', 'max', 'mean', 'std'], 'Consumer Confidence Index':['min', 'max', 'mean', 'std']}).plot(kind='barh')
plt.yticks(rotation=45)
plt.title('Relationship between Inflation Rate and Consumer Confidence Index');
```
